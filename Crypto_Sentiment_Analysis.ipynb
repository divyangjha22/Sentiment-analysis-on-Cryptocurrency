```python
# Seperating the article publish date & time
df['publishedAt'] = pd.to_datetime(df['publishedAt'])
df['Date'] = df['publishedAt'].dt.date
df['Date'] = df['publishedAt'].dt.strftime('%Y-%m-%d')
df['Time'] = df['publishedAt'].dt.time
df
```

```python
# Converting the time datatype (ie. datetime) into string so later we can perform vizualization on Time
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Time_str'] = df['Time'].apply(lambda x: x.strftime('%H:%M:%S'))
df
```

```python
df=df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python
df1=df[['Date','Time_str','author','title','description','url']]
df1
```

## Polarity

- Polarity determines if the text expresses the positive, negative or neutral sentiment of the text.
- It range the sentiment between -1 to 1.

```python
# Function for finding the polarity of the text
def getpolarity(text):
    return TextBlob(text).sentiment.polarity

# Creating new column for polarity
df1['Polarity']= df['title'].apply(getpolarity)
```

```python
df1.head()
```

```python
# Function to check if the sentiments of the title is posative, negative or neutral.
def getsentiment(score):
    if score<0:
        return 'Negative'
    elif score==0:
        return 'Neutral'
    else:
        return 'Posative'
    
# New column to store the sentiments
df1['Sentiments']=df1['Polarity'].apply(getsentiment)
```

```python
df1.head()
```

```python
# Renaming the columns name
df1.rename(columns={'author':'Author_Name','title':'Headlines','description':'Description','url':'URL'},inplace=True)
df1.head()
```

```python
# Performing vizualization of sentiments for better undestanding
plt.title('Sentiment Analysis')
df1['Sentiments'].value_counts().plot(kind='bar')
plt.xlabel('Sentiments')
plt.ylabel('Count')
plt.show()
```

```python
plt.title('Sentiment Analysis')
df1['Sentiments'].value_counts().plot(kind='pie')
plt.xlabel('Sentiments')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the sum of each polarity in terms of time
plt.figure(figsize=(100,5))
plt.title('Sentiments sum over the time')
plt.xlabel('Time')
plt.ylabel('Snetiments')
polarity=df1.groupby(['Time_str']).sum()['Polarity']
plt.plot(polarity.index,polarity)
```

```python
polarity
```

```python
polarity_count=df1.groupby(['Time_str']).count()['Polarity']
polarity_count
```

```python
# Average Sentiment
polarity_avg=polarity/polarity_count
polarity_avg
```

```python
# Plotting average sentiments over time
plt.figure(figsize=(100,5))
plt.title('Sentiments average over time')
plt.xlabel('Time')
plt.ylabel('Snetiments')
plt.plot(polarity_avg.index,polarity_avg)
```
